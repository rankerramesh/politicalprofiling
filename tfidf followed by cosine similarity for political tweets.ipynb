```python
wordSet=set(bowCongress).union(set(bowUML)).union(set(bowMao)).union(set(bowMadhesi)).union(set(bowSajha)).union(set(bowRaprapa)).union(set(bowNonPolitical))
```

```python
wordDictCongress=dict.fromkeys(wordSet,0)
wordDictUml=dict.fromkeys(wordSet,0)
wordDictMao=dict.fromkeys(wordSet,0)
wordDictMadhesi=dict.fromkeys(wordSet,0)
wordDictSajha=dict.fromkeys(wordSet,0)
wordDictRaprapa=dict.fromkeys(wordSet,0)
wordDictNonPolitical=dict.fromkeys(wordSet,0)
```

```python
def computeTf(wordDict,bow):
    tfDict={}
    bowCount=len(bow)
    for word, count in wordDict.items():
        tfDict[word]=count/float(bowCount)
    return tfDict
```

```python
tfBowCongress=computeTf(wordDictCongress,bowCongress)
tfBowUml=computeTf(wordDictUml,bowUML)
tfBowMao=computeTf(wordDictMao,bowMao)
tfBowMadhesi=computeTf(wordDictMadhesi,bowMadhesi)
tfBowSajha=computeTf(wordDictSajha,bowSajha)
tfBowRaprapa=computeTf(wordDictRaprapa,bowRaprapa)
tfBowNonPolitical=computeTf(wordDictNonPolitical,bowNonPolitical)
```

```python
pd.DataFrame([tfBowCongress,tfBowUml,tfBowMao,tfBowMadhesi,tfBowSajha,tfBowRaprapa,tfBowNonPolitical])
```

```python
def removekey(d, key):
    r = dict(d)
    del r[key]
    return r 
def computeIDF(docList):
    import math
    idfDict={}
    N=len(docList)
    #count the number of documents that contains the word w
    idfDict=dict.fromkeys(docList[0].keys(),0)
    for doc in docList:
        for word,val in doc.items():
            if val>0:
                idfDict[word]+=1
               
    #remove words with max 80% of doc     and 20 %            
    #divide N by numerator abov and take log
    for word, val in idfDict.items():
        if(int(idfDict[word])>5 or int(idfDict[word])<2):
            removekey(idfDict,word)
            
    for word, val in idfDict.items():
        idfDict[word]=math.log(N/float(val))
    return idfDict
```

```python
idfs=computeIDF([wordDictCongress,wordDictUml,wordDictMao,wordDictMadhesi,wordDictSajha,wordDictRaprapa,wordDictNonPolitical])
```

```python
def computeTFIDF(tfBow,idfs):
    tfidf={}
    for word,val in tfBow.items():
        tfidf[word]=val*idfs[word]
    return tfidf
```

```python
tfidfBowCongress=computeTFIDF(tfBowCongress,idfs)
tfidfBowUml=computeTFIDF(tfBowUml,idfs)
tfidfBowMao=computeTFIDF(tfBowMao,idfs)
tfidfBowMadhesi=computeTFIDF(tfBowMadhesi,idfs)
tfidfBowSajha=computeTFIDF(tfBowSajha,idfs)
tfidfBowRaprapa=computeTFIDF(tfBowRaprapa,idfs)
tfidfBowNonPolitical=computeTFIDF(tfBowNonPolitical,idfs)
```

```python
dfAll=pd.DataFrame([tfidfBowCongress,tfidfBowUml,tfidfBowMao,tfidfBowMadhesi,tfidfBowSajha,tfidfBowRaprapa,tfidfBowNonPolitical])

```
